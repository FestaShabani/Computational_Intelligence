DEPTH LIMITED SEARCH

The code you provided is a Python implementation of Depth-Limited Search (DLS). Here's a breakdown of how the code works:

The depth_limited_search function is the entry point for starting the DLS from the given node to the goal with the specified depth_limit.

The recursive_dls function is the recursive part of the search. It explores the graph depth-first up to a certain depth limit. It has the following logic:

If the current node is the goal, it immediately returns a list with just that node to indicate that the goal has been found.
If the depth_limit is 0, it returns an empty list because it has reached the depth limit without finding the goal.
If the depth_limit is greater than 0, it iterates through the neighbors of the current node. For each neighbor, it recursively calls recursive_dls with a reduced depth_limit. If a non-empty result is obtained from the recursive call, it means a path to the goal has been found, and the function returns a list starting with the current node and followed by the rest of the path.
The depth_limited_search function returns the result of the recursive DLS search, which is either an empty list (if no path was found) or a list representing the path from the node to the goal.

```python
def depth_limited_search(graph, node, goal, depth_limit):
    return recursive_dls(graph, node, goal, depth_limit)

def recursive_dls(graph, node, goal, depth_limit):
    if node == goal:
        return [node]

    if depth_limit == 0:
        return []

    if depth_limit > 0:
        print("depth_limit", depth_limit)
        for neighbor, _ in graph.get(node, []):
            print("FOR neighbor,_" ,neighbor,_)
            print("of this node->",node, [])
            print("IN neighbors of that node",graph.get(node, []) )
            print("recurse now\n")
            result = recursive_dls(graph, neighbor, goal, depth_limit - 1)
            if result:
                print("result", result)
                print("[node] + result",[node] + result)
                return [node] + result

    return []

start_node = 'A'
goal_node = 'M' #tried H before
depth_limit = 3

path = depth_limited_search(graph, start_node, goal_node, depth_limit)
if path:
    print("Path found:", ' -> '.join(path))
else:
    print("Path not found within the depth limit.")
```

BREADTH FIRST SEARCH

```python
from collections import deque

def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    while queue:
        (node, path) = queue.popleft()  #the differnce is here from DFS
        if node == goal:
            return path
        for (neighbor, _) in graph[node]:
            if neighbor not in path:
                queue.append((neighbor, path + [neighbor]))
    return []

result_bfs = bfs(graph, 'A', 'E')
print("BFS Path:", result_bfs)
```

UNIFORM COST SEARCH

The pq (priority queue) is used to store elements as tuples, where each tuple has three elements:

current cost: The cumulative cost from the start node to the current node.
current node: The current node being considered in the search.
path from start node to the current node: A list representing the path from the start node to the current node.
The search loop continues while the priority queue is not empty.

In each iteration, the algorithm retrieves the element with the lowest cost from the priority queue using heapq.heappop(pq).

If the current node is the goal node, the algorithm returns the path found so far, which is the shortest path from the start node to the goal node.

If the current node is not the goal, the algorithm explores its neighbors. For each neighbor that hasn't been visited yet (i.e., not in the path), the algorithm calculates the total cost from the start node to that neighbor, and it pushes the neighbor, the updated cost, and the updated path into the priority queue.

The algorithm repeats these steps until it reaches the goal node or exhausts all possible paths.

```python
import heapq

def uniform_cost_search(graph, start, goal):
    pq = [(0, start, [start])] #priority queue that stores elements as tuples.
    #pq = [current cost, current node, path from start node to the current node]
    print("start pq",pq)
    while pq:
        (cost, node, path) = heapq.heappop(pq) #uses a heap, # Removes and returns the smallest element from the heap.
        print("# least costly node (cost, node, path)", (cost, node, path))
        if node == goal:
            return path
        for (neighbor, edge_cost) in graph[node]: #graph[node] is node's neighbours and its costs
            print("#(neighbor, edge_cost)",(neighbor, edge_cost))
            print("#neighbours of that node ", graph[node])
            if neighbor not in path:
                heapq.heappush(pq, (cost + edge_cost, neighbor, path + [neighbor]))
                print("neighbour pushed in #heapq because not in path: cost + edge_cost, neighbor, path + [neighbor] ", cost + edge_cost, neighbor, path + [neighbor])
                print(".")
        print("\n")
    return []

result_uniform = uniform_cost_search(graph, 'A', 'N')
print("Uniform Cost Search Path:", result_uniform)
```
